### Imports

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mticks
import matplotlib.dates as mdates
```

## Prepare the dataset for time series analysis

### Load in the Zillow Home Value Index dataframe

```python
df = pd.read_csv('https://docs.google.com/spreadsheets/d/e/2PACX-1vQN3Ksa9szQuO4G1-msXWAp17KtVHQCBnuEieu_auu1wSiBf3-krHusIx5VBMkihxj-KZLBosDIGEyR/pub?output=csv')
pd.set_option("display.max_columns", 300)
display(df.head())
display(df.info())
```

### Filter the 4 largest cities into a new dataframe

```python
df_largest = df.head(4)
df_largest
```

### Melt the data to long-form and prepare it for time series analysis

```python
id_cols = ['RegionID', 'SizeRank', 'RegionName', 'RegionType', 'StateName', 'State', 'Metro', 'CountyName']
melted = pd.melt(df_largest, id_vars = id_cols, value_name = 'House Value', var_name = 'Month')
melted
```

### Convert the melted dates into datetime datatype

```python
melted['Month'] = pd.to_datetime(melted['Month'])
melted.info()
```

### Make the datetime column the index

```python
melted = melted.set_index('Month')
melted.head()
```

### Resample the dataframe as monthly frequency, grouped by City

```python
df_ts = melted.groupby('RegionName').resample("M").mean()
df_ts
```

## Plot the home values for all 4 cities

```python
cities = ['New York', 'Los Angeles', 'Houston', 'Chicago']
plot_df = df_ts.loc[cities].copy()
plot_df = plot_df.drop(columns = ['RegionID', 'SizeRank'])
plot_df
```

```python
plot_df.index.get_level_values(0)
```

```python
plot_df.index.get_level_values(1)
```

```python
unstacked = plot_df.unstack(level=0)
unstacked
```

```python
# create a wide figure
plt.rcParams['figure.figsize'] = (12,3)
```

```python
# create tick locators and formatters
loc_year = mdates.YearLocator(3)
fmt_year = mdates.DateFormatter("%Y")
loc_3months = mdates.MonthLocator(interval = 3)
```

```python
# function for formatting money in thousands
def thousands(x, pos):
    new_x = x/1_000
    return f"{new_x:,.0f}K"
fmt_thous = mticks.FuncFormatter(thousands)
```

```python
fig, ax = plt.subplots()
ax.plot(unstacked.index, unstacked.values, label = unstacked['House Value'].columns)
ax.set(title = 'House Values',xlabel = 'Date', ylabel = 'Thousands of Dollars')
ax.xaxis.set_major_locator(loc_year)
ax.xaxis.set_major_formatter(fmt_year)
ax.xaxis.set_minor_locator(loc_3months)
ax.yaxis.set_major_formatter(fmt_thous)
ax.legend();
fig = ax.get_figure()
fig.autofmt_xdate(rotation = 90, ha = 'center')
```

## Which City had the highest Typical Home Value at the end of 2008? Which had the least?

```python
unstacked.loc[pd.IndexSlice['2008-12-31']].sort_values()
```

- New York had the highest typical home value at the end of 2008.
- Houston had the lowest typical home value at the end of 2008.

## How much did the home values change from November 2008 to December 2008 (in dollars)?

```python
unstacked_diff = unstacked.diff()
unstacked_diff.loc[pd.IndexSlice['2008-12-31']]
```

- Home values in Chicago dropped by 5753.
- Home values in Houston dropped by 964.
- Home values in Los Angeles dropped by 12,611.
- Home values in New York dropped by 4458.
