## Prepare the dataset for time series analysis

### Load in the Zillow Home Value Index dataframe

```python
df = pd.read_csv('https://docs.google.com/spreadsheets/d/e/2PACX-1vQN3Ksa9szQuO4G1-msXWAp17KtVHQCBnuEieu_auu1wSiBf3-krHusIx5VBMkihxj-KZLBosDIGEyR/pub?output=csv')
pd.set_option("display.max_columns", 500)
display(df.head())
display(df.info())
```

### Filter the 4 largest cities into a new dataframe

```python
df_largest = df.head(4)
df_largest
```

### Melt the data to long-form and prepare it for time series analysis

```python
id_cols = ['RegionID', 'SizeRank', 'RegionName', 'RegionType', 'StateName', 'State', 'Metro', 'CountyName']
melted = pd.melt(df_largest, id_vars = id_cols, value_name = 'House Value', var_name = 'Month')
melted
```

### Convert the melted dates into datetime datatype

```python
melted['Month'] = pd.to_datetime(melted['Month'])
melted.info()
```

### Make the datetime column the index

```python
melted = melted.set_index('Month')
melted.head()
```

### Resample the dataframe as monthly frequency, grouped by City

```python
df_ts = melted.groupby('RegionName').resample("M").mean()
df_ts
```
